```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from statistics import variance 
df = pd.read_csv('../Wave Forms/test.csv')
```

```python

plt.plot(df)
```

```python
data = df[df.columns[0]].values.tolist()
T = len(data)
aic_data = []

def AIC(t):
    aic = t* np.log10(variance(data[0:t])) + (T-t)* np.log10(variance(data[t:T]))
    return aic 

for t , xt in enumerate(data):

    if t not in [0,1,T-1,T]:
        aic_data.append(AIC(t))
        
```

```python

%matplotlib inline
plt.plot(aic_data)
```

```python
from matplotlib import rc
rc('mathtext', default='regular')

fig = plt.figure()
ax = fig.add_subplot(111)
ax2 = ax.twinx()

p1 , = ax.plot(aic_data, color='black' , label = "AIC function")
p2 , = ax2.plot(data[3:], color='red' , label = "signal")
p3 , = ax.plot(aic_data.index(np.amin(aic_data)), np.amin(aic_data), marker="o", markersize=5, markerfacecolor="green" , label="AIC_min")
ax.legend(handles=[p1, p2,p3])
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib import rc
from statistics import variance 
import json 

rc('mathtext', default='regular')

sensor_num = 1
hit_num = 1
df = pd.read_csv(f'../Wave Forms/hit_{hit_num}/sensor{sensor_num}.csv')

data = df[df.columns[0]].values.tolist()

def AIC(data):

    aic_data = []
    T = len(data)

    for t in range(T):

        if t not in [0,1,T-1 ,T] :
            #print(t)
            aic = t* np.log10(variance(data[0:t])) + (T-t-1)* np.log10(variance(data[t:T]))        
            aic_data.append(aic)

    return aic_data

aic_data = AIC(data)



```

```python
figure, ax = plt.subplots(2, 2)
ax2 = ax[0, 0].twinx()

p1 , = ax[0,0].plot(aic_data, color='black' , label = "AIC function")
p2 , = ax2.plot(data[3:], color='red' , label = "signal")
p3 , = ax[0,0].plot(aic_data.index(np.amin(aic_data)), np.amin(aic_data), marker="o", markersize=5, markerfacecolor="green" , label="AIC_min")
ax[0,0].legend(handles=[p1, p2,p3])
# ax.set_ylabel("AIC(t)" ,fontsize = 13 )
# ax2.set_ylabel("X(t)" ,fontsize = 13 )
# ax.set_xlabel("t" ,fontsize = 13)
plt.title(f"Hit {hit_num} - Sensor {sensor_num} Data" , fontsize = 15)
plt.show()
```
